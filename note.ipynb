```python
import pandas as pd

# Load the data from both CSV files
df1 = pd.read_csv('ens.csv')
df2 = pd.read_csv('ensemble_submission.csv')

# Assuming both files have the same ordering of rows and same 'inst_id'
same_oc = (df1['OC'] == df2['OC']).sum()
different_oc = len(df1) - same_oc

print(f"Number of same 'OC' entries: {same_oc}")
print(f"Number of different 'OC' entries: {different_oc}")
```
